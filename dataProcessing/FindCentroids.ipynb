<h2>Find coordinates of census track centers and add 'hover' data</h2>

```python
from geojson_utils import point_in_multipolygon
from geojson_utils import centroid
import pandas as pd
import numpy as np
import json
```

```python
data= pd.read_json('C:/Users/mskac/machineLearning/GasLeakConEd/data/2010_census_tracks_with_geoid.geojson')
```

```python
centers = pd.DataFrame(columns=['geoid', 'boro', 'nta', 'centerLong', 'centerLat'])
for element in data['features']:
    box_str = str(element['geometry']['coordinates'][0])
    box_str = '{"type": "Polygon","coordinates": '+box_str+'}'
    box = json.loads(box_str)
    point = centroid(box)
    centers = centers.append({'geoid': element['properties']['geoid'], 'nta': element['properties']['ntaname'].lower(), 'boro': element['properties']['boro_name'].lower(), 'centerLong': point['coordinates'][0], 'centerLat': point['coordinates'][1]}, ignore_index=True)
```
